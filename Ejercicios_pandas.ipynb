```python
fechas = ['2000-2005','2006-2010','2011-2015','2016-2020','2021-2025',]
ventas=[]
for rango in fechas:
  a = int(input(f"Ingrese la venta en el rango de {rango}: "))
  ventas.append(a)
Serie = pd.Series(ventas,index=fechas)
print("\nVentas originales:")
print(Serie)
serieDescuento = Serie * 0.9
print("\nVentas con 10% de descuento:")
print(serieDescuento)
```

Ejercicio 2

```python
def estadisticas_notas(dic_notas):
    notas = pd.Series(dic_notas)
    resumen = pd.Series({
        'Nota mínima': notas.min(),
        'Nota máxima': notas.max(),
        'Nota media': notas.mean(),
        'Desviación típica': notas.std()
    })
    return resumen
notas = {
    'Juan': 15,
    'María': 18,
    'Luis': 13,
    'Ana': 17
}

resultado = estadisticas_notas(notas)
print(resultado)
```

Ejercicio 3

```python
def obtener_aprobados(dicNotas):
    nota = pd.Series(dicNotas)
    aprobados = nota[nota >= 11]
    ordenados = aprobados.sort_values(ascending=False)
    return ordenados

notas_alumnos = {
    'Juan': 8,
    'María': 14,
    'Luis': 11,
    'Ana': 17,
    'Pedro': 9
}

resultado = obtener_aprobados(notas_alumnos)
print(resultado)
```

```python
survival_counts = df['Survived'].value_counts(normalize=True) * 100
print("Porcentaje que murieron (0):", round(survival_counts[0], 2), "%")
print("Porcentaje que sobrevivieron (1):", round(survival_counts[1], 2), "%")
```

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
for pclass, percentage in survival_by_class.items():
    print(f"Clase {pclass}: {round(percentage, 2)}% sobrevivieron")
```

```python
df.dropna(subset=['Age'])
```

```python
edad_media_mujeres = df[df['Sex'] == 'female'].groupby('Pclass')['Age'].mean()
print("Edad media de mujeres por clase:")
print(edad_media_mujeres)
```

```python
df['MenorDeEdad'] = df['Age'] < 18
df
```

```python
supervivencia_menores = df.groupby(['Pclass', 'MenorDeEdad'])['Survived'].mean() * 100
print("Porcentaje de supervivencia por clase y edad:")
for (pclass, menor), porcentaje in supervivencia_menores.items():
    tipo = "Menor" if menor else "Mayor"
    print(f"Clase {pclass} - {tipo} de edad: {round(porcentaje, 2)}% sobrevivieron")
```
